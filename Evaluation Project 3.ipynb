```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
```

```python
maxAcc=0
maxRS=0
for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=i)
    model=DecisionTreeRegressor()
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    score=model.score(x_train,y_train)
    if score>maxAcc:
        maxAcc=score
        maxRS=i
print('Best score is', maxAcc,'on random state',maxRS)        
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(lm,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(dtr,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(svr,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(rf,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(knn,x,y,cv=5)
print(score)
print(score.mean())
```

# Hyper Parameter Tuning

We use this technique to improve the accuracy of the best model that we have choosen.

```python
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[85,100,115],
            'criterion':['mse','mae'],
            'max_depth':range(2,10,2),
            'max_features':['auto','sqrt','log2']}
gs=GridSearchCV(RandomForestRegressor(),parameters,cv=5)
gs.fit(x_train,y_train)
```

```python
print(gs.best_params_)
```

```python
rfmodel=RandomForestRegressor(criterion='mse',max_depth=8,max_features='log2',n_estimators=115)
rfmodel.fit(x_train,y_train)
score=rfmodel.score(x_train,y_train)
print(score*100)
```
